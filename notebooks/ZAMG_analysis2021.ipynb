### Zusammenfassung

Datengrundlage
- Daten ZAMG Wetterstation Weitra (Station 1800 bzw. 1802 (ab 2003)), https://data.hub.zamg.ac.at/
- Daten sind erst ab 1936 verfügbar, aufgrund fehlender Datenpunkte mussten gewisse Jahre ausgeschlossen werden
- Analyse wurde für den Zeitraum 1936-2020 gemacht
- Für Trendanalysen wurde der Zeitraum in 3 Teilblöcke unterteilt: 1936-70, 1971-90, 1991-2020
- Daten scheinen von guter Qualität, Messfehler können aber nicht ausgeschlossen werden

Temperatur:
- deutlicher Temperaturanstieg im Mittel u. Im Median: ~ 1,4° im Vgl. zu vor 1970
- ansteigender Trend seit 1990 beobachtbar: 0.7 bis 0.88° pro Dekade bei Ann. eines linearen Trends
- TOP 10 wärmste Jahre kommen alle aus den 2000ern, kälteste Jahre alle vor 1980
- Kältewellen mit Temperaturen unter -5° an mehreren Tagen hintereinander nehmen ab
- keine Temperaturen über 32 Grad vor 1990
- Hitzewellen werden volatiler weil länger als vor 1990

Niederschlag:
- die jährliche Gesamtmenge zeigt keinen Trend (die Periode 1971-1990 ist ein wenig schwächer)
- die mittlere Menge Regen an Regentagen nimmt aber ab
- Starkregenereignisse nehmen aber zu:
- Top 3 tägl. Regenmengen liegen alle in den 2000ern
- 5 von 10 in den Top den sind in den 2000ern gemessen worden
- Trockenperdioden: eigentlich keine Verlängerung der Perioden in denen es nicht regnet zu erkennen, eigentlich werden diese sogar kürzer

Schnee:
- sinkender Trend bei Schneemenge: sowohl bei Mittelwert als auch bei maximalen Tagesmengen, dieser Rückgang ist aber schon seit den 70ern zu beobachten
- absteigender Trend auch bei den Tagen wo Schneebedeckung zu messen ist, wobei es durchaus auch nach 2000 Jahre gibt wo sehr viele Schneetage beobachtbar waren, zb 2006 als Maximum der ganzen Periode
- Perioden mit durchgehender Schneebedeckung werden aber kürzer

Wind:
- im Mittel sogar abnehmend! -> Datenproblem?!
- bei Maximalgeschwindigkeit leider erst Beobachtungen ab 2003 verfügbar

Sonneeinstrahlung:
- Daten erst ab 80ern verfügbar
- hat aber im Mittel deutlich zugenommen: +1h pro Tag

```python
# merge data sets
data_total = pd.concat([data_till0309[data_till0309['time'] < '2003-09-01'], data_from0309[data_from0309['time']>='2003-09-01']], ignore_index = True)
```

```python
print('Grob-Analyse der Daten: \n')
data_total.describe()
```

```python
#Bereinigung von negativen Werten in Niederschlag und Schneebedeckung
data_clean = data_total
data_clean['nied_clean'] = data_clean['nied'].clip(0, None)
data_clean['schnee_clean'] = data_clean['schnee'].clip(0, None)

print('Analyse der bereinigten Daten für Niederschlag u. Schnee: \n')
data_clean.drop(['nied', 'schnee'], axis = 1).describe()
```

```python
# Analyse wann Daten beginnen
data_clean['jahr'] = pd.to_datetime(data_clean['time']).dt.year
data_clean['monat'] = pd.to_datetime(data_clean['time']).dt.month
data_clean['tag'] = pd.to_datetime(data_clean['time']).dt.day
check = data_clean.groupby(by = 'jahr').agg(anzahl=("nied_clean", "count")).reset_index()
#print(check[(0<check['anzahl']) and (check['anzahl'] <365)])
display(check[((check['anzahl']<365))].tail(6))
print('Daten erst ab 1936 verfügbar, Jahr 2011 bis November, -> Jahre 1945, 1951-1954 ausgeschlossen aus der Analyse da unvollständig!')
```

```python
data_final = data_clean[data_clean['jahr'] > 1936]
data_final = data_final[-data_final['jahr'].isin([1945,1951,1952,1953,1954,2021])].reset_index()# exclude incomplete years, 2000 march is also not fully complete -> neglected!
data_final = data_final.drop('index', axis = 1)
#print(data_final)
```

### Analyse tägliches Temperaturmittel 2m: t

```python
#print(data_final[data_final['nied_clean'].isna()==True])
```

```python
# aggregiert auf Jahresbasis - Mittelwert
t_yearly = data_final.groupby(by = 'jahr').aggregate(t_mittel_j = ('t', 'mean')).reset_index()
t_yearly['x_tick'] = t_yearly['jahr'].apply(lambda x: str(x) if x % 10 == 0 else '') # create ticks for decades

part1 = t_yearly[t_yearly['jahr']<1971]
part2 = t_yearly[t_yearly['jahr'].isin(range(1971,1991))]
part3 = t_yearly[t_yearly['jahr']>1990]

p1_avg = np.ones(len(part1['jahr'])) * part1['t_mittel_j'].mean()
p2_avg = np.ones(len(part2['jahr'])) * part2['t_mittel_j'].mean()
p3_avg = np.ones(len(part3['jahr'])) * part3['t_mittel_j'].mean()

fig, ax = plt.subplots(figsize=(30,10))
ax.plot(t_yearly['jahr'], t_yearly['t_mittel_j'], label = 'jährl. Durchschnittstemp.') # j. Schnitt
ax.plot(part1['jahr'], p1_avg, label = 'Schnitt 1936-1970')
ax.plot(part2['jahr'], p2_avg, label = 'Schnitt 1971-1990')
ax.plot(part3['jahr'], p3_avg, label = 'Schnitt 1991-2020')
ax.set_xlabel('Jahr')
ax.set_ylabel('Grad Celcius')
ax.set_xticks(ticks = np.arange(1930,2021,10))

ax.grid(which='major', color='#DDDDDD', linewidth=0.8)
ax.grid(which='minor', color='#EEEEEE', linestyle='--', linewidth=0.5)
ax.minorticks_on()

plt.title('Enwicklung Jährliche Durchschnittstemperatur 1936 - 2020')
plt.legend()
plt.show()
#print(t_yearly)
```

```python
# Top 5 wärmste u. kälteste Jahre u. Trend in den 90er Jahren
kalt = t_yearly.sort_values(by = 't_mittel_j', ascending = True)
warm = t_yearly.sort_values(by = 't_mittel_j', ascending = False)

print('wärmste Jahre Mittelwert, 1936 - 2020:')
display(warm.drop('x_tick', axis  =1 ).head(10))

print('\nkälteste Jahre Mittelwert, 1936 - 2020:')
display(kalt.drop('x_tick', axis  =1 ).head(10))

#calculate linear trend for part 3
Y = part3['t_mittel_j']
X = part3['jahr']
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()
trend10 = 10*model.params[1]
#print(model.params)

print(f'\nErwärmung innerhalb einer Dekade seit 1991 bei Annahme eines linearen Trends in 1991-2020: {trend10:.2f} Grad.')
```
